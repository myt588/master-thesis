```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cross_validation as cv
from sklearn.metrics.pairwise import pairwise_distances
```

```python
# Load Data
sentences_df = pd.read_csv('sentences.csv')
sentences_df = sentences_df[['sentence_id', 'flesch_kincaid']]
users_df = pd.read_csv('sentences-user-actions.csv')
users_df = users_df[['username', 'sentence_id', 'unknown_count']]

result = pd.merge(sentences_df, users_df, on='sentence_id', how='inner')

a = result.groupby('flesch_kincaid', as_index=False)['unknown_count'].sum()

b = pd.merge(a, result, on='flesch_kincaid', how='inner')

# individual user plot
for x in range(0, 36):
    c = b[b['username'] == x]
    plt.scatter(c['unknown_count_x'], c['unknown_count_y'])
    plt.show()
```

```python
n_users = users_df['username'].unique().shape[0]
n_items = sentences_df['sentence_id'].unique().shape[0]
print('Number of users = ' + str(n_users) + ' | Number of readings = ' + str(n_items))

def predict(ratings, similarity, type='user'):
    if type == 'user':
        pred = similarity.dot(ratings) / np.array([np.abs(similarity).sum(axis=1)]).T
    elif type == 'item':
        pred = ratings.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
    return pred

from sklearn.metrics import mean_squared_error
from math import sqrt
def rmseWithoutZeors(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten()
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, ground_truth))

def rmse(prediction, ground_truth):
    prediction = prediction.flatten()
    ground_truth = ground_truth.flatten()
    return sqrt(mean_squared_error(prediction, ground_truth))

def percentCorrect(prediction, ground_truth, bound):
    prediction = prediction[ground_truth.nonzero()].flatten()
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    in_bound = 0
    for index, x in enumerate(prediction):
        if (x <= ground_truth[index] + bound and x >= ground_truth[index] - bound):
            in_bound = in_bound + 1
    return in_bound/len(prediction)
    
user_prediction_rmse_s_sum = 0
item_prediction_rmse_s_sum = 0
user_prediction_rmse_sum = 0
item_prediction_rmse_sum = 0
user_percent_correct_sum = 0
item_percent_correct_sum = 0
training_repeat = 10
for x in range(0, training_repeat):
    train_data, test_data = cv.train_test_split(result, test_size=0.25)

    train_data_matrix = np.zeros((n_users, n_items))
    for line in train_data.itertuples():
        train_data_matrix[line[3]-1, line[1]-1] = line[4]

    test_data_matrix = np.zeros((n_users, n_items))
    for line in test_data.itertuples():
        test_data_matrix[line[3]-1, line[1]-1] = line[4]

    # calculate similarity
    user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
    item_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')

    # predict
    user_prediction = predict(train_data_matrix, user_similarity, type='user')
    item_prediction = predict(train_data_matrix, item_similarity, type='item')

    user_prediction_rmse_s = rmseWithoutZeors(user_prediction, test_data_matrix)
    user_prediction_rmse_s_sum = user_prediction_rmse_s_sum + user_prediction_rmse_s
    item_prediction_rmse_s = rmseWithoutZeors(item_prediction, test_data_matrix)
    item_prediction_rmse_s_sum = item_prediction_rmse_s_sum + item_prediction_rmse_s
    print('Sentence-wise User-based CF RMSE: ' + str(user_prediction_rmse_s))
    print('Sentence-wise Item-based CF RMSE: ' + str(item_prediction_rmse_s))
    
    user_percent_correct = percentCorrect(user_prediction, test_data_matrix, 2)
    user_percent_correct_sum = user_percent_correct_sum + user_percent_correct
    item_percent_correct = percentCorrect(item_prediction, test_data_matrix, 2)
    item_percent_correct_sum = item_percent_correct_sum + item_percent_correct
    print('Percent Correct User-based: ' + str(user_percent_correct))
    print('Percent Correct Item-based: ' + str(item_percent_correct))

    
print('Sentence-wise User-based CF RMSE Avg: ' + str(user_prediction_rmse_s_sum/training_repeat))
print('Sentence-wise Item-based CF RMSE Avg: ' + str(item_prediction_rmse_s_sum/training_repeat))
print('Sentence-wise User-based CF Correct Percent Avg: ' + str(user_percent_correct_sum/training_repeat))
print('Sentence-wise Item-based CF Correct Percent Avg: ' + str(item_percent_correct_sum/training_repeat))
```

```python
def predictionSum(user_prediction, item_prediction, test_data_matrix):
    # sum up count in sentences to readings
    user_prediction_sum = np.zeros((n_users, 7))  
    index = 0
    for line in user_prediction:
        user_prediction_sum[index, 0] = line[:12].sum()
        user_prediction_sum[index, 1] = line[13:17].sum()
        user_prediction_sum[index, 2] = line[18:40].sum()
        user_prediction_sum[index, 3] = line[41:64].sum()
        user_prediction_sum[index, 4] = line[65:94].sum()
        user_prediction_sum[index, 5] = line[95:107].sum()
        user_prediction_sum[index, 6] = line[108:137].sum()
        index = index + 1

    item_prediction_sum = np.zeros((n_users, 7))  
    index = 0
    for line in item_prediction:
        item_prediction_sum[index, 0] = line[:12].sum()
        item_prediction_sum[index, 1] = line[13:17].sum()
        item_prediction_sum[index, 2] = line[18:40].sum()
        item_prediction_sum[index, 3] = line[41:64].sum()
        item_prediction_sum[index, 4] = line[65:94].sum()
        item_prediction_sum[index, 5] = line[95:107].sum()
        item_prediction_sum[index, 6] = line[108:137].sum()
        index = index + 1

    index = 0
    truth_sum = np.zeros((n_users, 7))
    for line in test_data_matrix:
        truth_sum[index, 0] = line[:12].sum()
        truth_sum[index, 1] = line[13:17].sum()
        truth_sum[index, 2] = line[18:40].sum()
        truth_sum[index, 3] = line[41:64].sum()
        truth_sum[index, 4] = line[65:94].sum()
        truth_sum[index, 5] = line[95:107].sum()
        truth_sum[index, 6] = line[108:137].sum()
        index = index + 1
            
    user_prediction_rmse = rmse(user_prediction_sum, truth_sum)
    user_prediction_rmse_sum = user_prediction_rmse_sum + user_prediction_rmse
    item_prediction_rmse = rmse(item_prediction_sum, truth_sum)
    item_prediction_rmse_sum = item_prediction_rmse_sum + item_prediction_rmse
    print('Full User-based CF RMSE: ' + str(user_prediction_rmse))
    print('Full Item-based CF RMSE: ' + str(item_prediction_rmse))

print('Full User-based CF RMSE Avg: ' + str(user_prediction_rmse_sum/training_repeat))
print('Full Item-based CF RMSE Avg: ' + str(item_prediction_rmse_sum/training_repeat))
```
